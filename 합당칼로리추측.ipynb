#데이터 훈련

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Data_set = pd.read_csv('Training_data.csv')

x = Data_set.iloc[:,0:3]
y = Data_set.iloc[:,3]
```

```python
model = Sequential()
model.add(Dense(10,input_dim = 3, activation ='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
model.fit(x,y,epochs=1500)
```

#데이터 테스트

```python
test_x = np.array([[129,26,19]])
predictions = model.predict(test_x)
print("고칼로리 예상 확률은 %.1f%%입니다." % (predictions *100))
```

```python
test_x2 = np.array([[106,101,121]])
prediction2 = model.predict(test_x2)
print("고칼로리 예상 확률은 %.1f%%입니다." % (prediction2 *100))
```
